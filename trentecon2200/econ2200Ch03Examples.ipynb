```python
# process the download GDP table, and make a time series of real and nominal gdp
gdp = cansimpivot
gdpAnnual = gdp.resample('A').mean()
gdpAnnual09 = gdpAnnual.loc['2009-12-31':'2019-12-31', 'qGDP']
gdpAnnual09
```

```python
# Geometric mean annual growth rate of real GDP from 2009 to 2019
xbarG = (gdpAnnual09['2019-12-31'] / gdpAnnual09['2009-12-31'])**(1.0/10.0) - 1
print("Annual growth rate of real GDP over 2009-2019 is", xbarG)
```
