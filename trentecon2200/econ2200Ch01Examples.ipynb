#### Example: average consumption expenditure per household in 2017, provinces

```python
# Question: in 2017 what is the average annual expenditure per household in Canadian provinces?
# Method: calculate mean
from IPython.display import display, Math, Latex
display(Math(
r"""
\Large
\begin{align}
 & y = \frac{1}{10}\sum_{i=1}^{10} x_i, \quad \text{or} \quad y = \frac{x_1 + x_2 + ... + x_{10}}{n}
\end{align}
"""))
```

```python
# Data: Household spending, Canada, regions and provinces, Statistics Canada, Table ID 1110022201.
import pandas as pd
my_data = pd.read_csv("dataSources/1110022201-eng.csv",header=0, index_col=None).T
my_data.columns = ["Household Expenditure"]
my_data.head(10)
```

```python
# What do we learn from data?
y = (58364.0+54983.0+53732.0+52608.0+53969.0+66855.0+60624.0+65117.0+72957.0+71001.0) / 10.0
print(r'Average expenditure per household in Canada in 2017 is ')
print('y =', y )
```

```python
# Or use Python Pandas library
y = my_data.mean(0)
y
```
