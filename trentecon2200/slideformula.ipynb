- **Constant: $e = 2.71828$.**

- **Binomial probability mass distribution: $$P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}$$
 with mean $np$ and variance $np(1-p)$.**

- **Poisson probability mass distribution: $$P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}, \quad \text{for} \ k = 0, 1, 2, \cdots$$
mean $\lambda$ and variance $\lambda$.**
