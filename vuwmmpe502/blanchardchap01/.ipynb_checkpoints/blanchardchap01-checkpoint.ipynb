### 2. Economic tour of the world, three economies

* Economists often look at three main indicators in an economy (or country)

  - Output growth
  
  - Unemployment rate
  
  - Inflation rate

#### 2-1. The United States

* The United States is big

  - With an output of \$17.4 trillion in 2014, it accounted for 23\% of the world output.

* The U.S. standard of living is high

  - Output per capita is \$54,000, close to the highest in the world (Switzerland is one of top countries).



* The U.S. economy in 2015 was in decent shape, leaving much of the effects of the 2008-2009 crisis behind.

 <center><img src="pptimages/image8.png" width="600" height="300"/>  </center>

* But **interest rate is low.**
  - The federal funds rate--the interest rate the Federal Reserve controls--went from 2.5% in July 2007 to nearly 0% in December 2008.

* The U.S. Federal funds rate since 2000.

 <center><img src="pptimages/image9.png" width="600" height="250"/> </center>

* Why did the Federal Funds rate stop at zero?

  - This constraint is known as the **zero lower bound**

  - If it were negative, then everyone would hold cash rather than bonds.

* Why are low interest rates a potential issue?

  - Low interest rates limit the Fed’s ability to respond to **further** negative shocks.

  - Low interest rates may lead to excessive risk taking by investors to increase their returns.


* **Productivity growth has been slow.**

  - Productivity growth is important for a sustained increase in income per person, but since 2010, it has been only about half as it was in the 1990s.
  - The slowdown in productivity growth is worrisome because the standard of living especially for the poor may not increase.
 <center><img src="pptimages/image11.png" width="600" height="250"/> </center>

#### 2-2. The Euro Area
* The **European Union** (EU) is a group of 28 European counties with a common market.

* The **Euro area** is a group of 19 EU members that use one common currency, the _euro_.

* The Euro area faces two main issues today:
  - How to reduce unemployment?

  - How to function efficiently as a common currency area?

 <center><img src="pptimages/image12.png" width="600" height="250"/> </center>

* The Euro area under-performed in comparison with the U.S, in terms of recovery from recession.
  - Output growth in the Euro area remained close to zero between 2010 and 2014.

  - The unemployment rate was 11.1%. Also, lots of variation across member countries.

* Large variations in growth and unemployment within the Euro area raise debates on the benefits of common currency.


#### 2-3. China

* China has become increasingly important, largely because of its rapid growth and large population.
