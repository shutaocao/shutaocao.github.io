## What is the course about?

- Economics concepts and thinking, focusing on macroeconomics.

- Important macroeconomic variables and measurement.

- Macroeconomic **trends**

  - Long-run growth and its engine.

  - Selected markets: investment, financial.

  - The open economy.

- Macroeconomic **fluctuations**

  - Short-run aggregate demand and supply.
  
  - Business cycles, inflation, and unemployment.

  - Macroeconomic polices: fiscal policy and monetary policy.
