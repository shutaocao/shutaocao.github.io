## 1. Markets and Prices

- A **market** is an arrangement with which buyers and sellers to get information and trade with each other a particular good or service.

- Some markets are highly organized, e.g. stock market, commodity markets, where there is an auctioneer that helps set price and organize the sale.
  
- Many markets are less organized, e.g. ice-cream in Peterborough, corporate bond market. Labor market??

- A **competitive market** is a market that has many buyers and many sellers, each single buyer or seller has negligible influence the market price.

- A market is _perfectly competitive_ if no single buyer or seller has any influence on the market price. We often require that the good or service being sold in a perfectly competitive market is the same across all sellers.

- The money price of a good is the amount of money needed to buy it.

- The **relative price** of a good---the ratio of its money price to the money price of the next best alternative good---is its opportunity cost.

## 2. Demand

- The **quantity demanded** of any good is the amount of the good that buyers are willing to and able to purchase,  given a particular price.

- **Demand** for a good reflects

  1) wants of the good (preference, desire, wish);
  
  2) the ability to pay for an amount of the good.

### 2-1. The Law of Demand

- The **law of demand**: Other things remaining the same, the higher the price of a good, the smaller is the quantity demanded; the lower the price of a good, the larger is the quantity demanded.

- A **demand schedule** is a table that shows the relationship between the quantity demanded of a good and its price, _holding constant everything else_ that influences how much of the good consumers want to buy.

- The graph that shows the relationship between the quantity demanded and price is called the **demand curve**.

![alt](ParkinMacroIL/ParkinMacro_Ch03/image8.png "Demand schedule")

#### Willingness to pay

- The demand curve is also a curve showing the degree of willingness to pay as price varies.

- The smaller the quantity available, the higher is the price that someone is willing to pay for another unit.

- Willingness to pay measures **marginal benefit**.

![alt](ParkinMacroIL/ParkinMacro_Ch03/image10.png "Willingness to pay")

#### Why is the demand curve downward slopped?

- When price rises, demand falls. Why?

- Two effects:

  - Substitute effect: when the price of a good rises, substitutes may be more appealing. People switch between goods when prices change.
  
  - Income effect: when the price of a good rises, consumers can no longer afford to a same amount, they reduce the quantity purchased.
  
- The two effects will be studied extensively in microeconomics.

### 2-2. Changes in demand (, in fact the demand curve)

- Quantity demanded varies when price varies, holding constant _everything else_ that influences demand.

- When "everything else" changes, demand also changes, such change is a **shift in demand**.

- Demand curve shifts to the right, when influencing factors causes demand to increase.

- Demand curve shifts to the left, when influencing factors lead to a lower demand.

**Shifts in the demand (curve) versus movements along the demand curve**

![alt](ParkinMacroIL/ParkinMacro_Ch03/image13.png "Changes in demand")

#### Factors that change demand (in fact the demand curve)

- Income 

- Preferences (Tastes)

- The prices of related goods 

- Expectations
  - Expected future prices
  
  - Expected future income and credit

- Number of buyers

#### Income

- Does an increase in income lead to a higher demand for a good or service? It depends.

- A **normal good** is one for which demand increases as income increases.

- An **inferior good** is a good for which demand decreases as income increases.

- When income increases, consumers buy more of a normal good and the demand curve shifts rightward.

- When income increases, consumers buy less of an inferior good and the demand curve shifts leftward.

#### Preferences (or Tastes)

- If I like vanilla ice-cream more, my demand for it increases, demand curve for vanilla ice-cream shifts rightward.

- Economists normally do not explain changes in tastes. We do examine the impact of changes in tastes.

#### Price of related goods

- Two goods are **substitutes** if a fall in the prices of one good _reduces_ the demand for the other good.
  - Two goods are substitutes if they can be used in place of each other.

  - Example, hamburgers and sandwiches; pick-ups and sedans; iphones and Android-based cell phones.

- Two goods are **complements** if a fall in the prices of one good _raises_ the demand for the other good.
  - A complement is a good that is used in conjunction with another good.

  - Example, computers and software; gasoline and automobiles. 
  
- Robots and human workers: are they substitutes or complements?

#### Expectations
- Changes in the expectations regarding the future may affect the demand of good today.

  - An expected increase in future income may raise the demand for a normal good today.
  
  - An expected price increase in the future may raise the demand for good today.

#### Number of buyers
- **Market demand** is the sum of all the individual demands for a particular good or service.

- When there are more buyers of a particular good, market demand for the good rises.

## 3. Supply

- The **quantity supplied** of a good or service is the amount that producers are willing to and are able to sell.

- The **law of supply**: Other things remaining the same, the higher the price of a good, the greater is the quantity supplied; and the lower the price of a good, the smaller is the quantity supplied.

- The term **supply** refers to the entire relationship between the quantity supplied and the price of a good.

- The **supply curve** shows the relationship between the quantity supplied of a good and its price when all other influences on producers' sales remain the same.

- The **market supply** is the sum horizontally of all individual supplies.

![Willingness to supply](ParkinMacroIL/ParkinMacro_Ch03/image15.png "Willingness to supply")

### 3-1. Price and marginal cost
- Recall, the marginal cost of producing a good or service increase as the quantity produced increases.

- Producers are willing to supply a good only if price of the good allow them to cover the marginal cost of production.

- As the quantity produced increases, marginal cost increases, the lowest price at which the producer is willing to sell an additional unit will rise.

- This lowest price equals marginal cost, below which the producer stops supplying an additional unit of good.

### 3-2.  Changes in supply

- Quantity supplied varies when price varies, holding constant all other factors.

- When factors other than price change, the supply of a good also change, such change is a **shift in supply**.

- When factors other than price lead to increases in supply, the supply curve shifts rightward.

- When factors other than price lead to declines in supply, the supply curve shifts leftward.

![Shifts in supply](ParkinMacroIL/ParkinMacro_Ch03/image19.png "Shifts in supply")

#### Factors shifting the supply curve

- Prices of factors of production

- Production technology

- Prices of related goods?

- Expectations
  - regarding prices in the future
  
- The number of suppliers

- State of nature

#### Factor prices

- Recall, factors of production include labor and capital.

- If factor prices increase, supply falls, i.e. supply curve shifts to the left.

- Production also involves other inputs such as raw materials.

- When input prices increase, supply curve also shifts leftward.

#### Technology

- The technology for turning inputs into a good also determines supply.

- We consider that an improvement in technology means the same amount of factors (say labor) can produce more of a good.

- Current debates: Concerns about automation and machine learning replacing human workers.

#### Expectations

- If producers expect the price of a good to rise in the future, they may reduced production today and produce more in the future. This leads to a leftward shift of supply curve.

![Shifts in supply](ParkinMacroIL/ParkinMacro_Ch03/image17.png "Shifts in supply")

#### Prices of related goods

- According to textbook, but not standard (to my knowledge).

- A *substitute in production* for a good is another good that can be produced using the same resources.

- The supply of a good increases if the price of a substitute in production falls.

- Goods are _complements in production_ if they must be produced together.

- The supply of a good increases if the price of a complement in production rises.

#### Number of producers

- Given a price level, the more the number of producers, the higher the market supply.

#### State of nature

- Natural disasters can reduce supply.

- Economics of climate change: currently topical.

## 4. Market equilibrium

- **Market equilibrium** is a situation in which price has reached a level at which quantity demanded equals quantity supplied. 

- Equilibrium is a point where the demand curve and supply curve intersect.

- The price at the intersection is the **equilibrium price**.

- The **equilibrium quantity** is the quantity bought and sold at the equilibrium price.

- Price regulates buying and selling plans: it is the instrument used by the "invisible hand".
