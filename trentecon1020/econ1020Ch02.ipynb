## 1. Production possibilities frontier and opportunity cost

### 1-1. Production possibilities frontier

- **Production possibilities frontier** (PPF) is a graph showing the various *combinations* of output that the economy can *possibly* produce (at most), given the available factors of production and the available production technology that firms can turn those factors into output.

- PPF highlights some basic ideas in economics: scarcity, efficiency, opportunity cost, trade-offs, economic growth.

- Note: Definition from the textbook is inaccurate. "Boundary between those combinations of goods and services that can be produced and those that cannot." Is boundary the frontier or not?

- To illustrate the PPF visually, we focus on two goods,  holding the quantities of all other goods and services constant.

- That is, we look at a model economy in which everything remains the same (ceteris paribus) except the two goods we're considering.

-Or, innocuously one can think that the economy produces only two goods.

![Production Possibilities Frontier](ParkinMacroIL/ParkinMacro_Ch02/image8.png "Production possibilities frontier")

![Production Possibilities Frontier](ParkinMacroIL/ParkinMacro_Ch02/image9.png "Production possibilities frontier")

#### Production efficiency

- **Production efficiency** is achieved if the economy cannot produce more of one good without producing less of some other good.

- All points on the PPF are efficient.

- Inside the PPF, some factors of production are idle.

### 1-2. Opportunity cost

- The production possibilities frontier shows the opportunity cost of one good as measured in terms of the other good.

- When the economy moves from point $C$ to point $D$, it gives up 3 million cokes to get 1 million pizzas. Roughly, moving from $C$ to $D$, the opportunity cost of each pizza is three cokes, the opportunity cost of each coke is 1/3 pizza.

- The opportunity cost of a pizza equals the _slope_ of the production possibilities frontier.

- The opportunity cost along the production possibilities frontier is **not** constant. This is reflected in the shape of the production possibilities frontier.

- Production possibilities frontier is bowed outward: suggesting that the opportunity cost of one pizza is the highest when the economy produces lots of pizzas and few cokes (say at point $E$), and the lowest when the economy produces few pizzas and lots of cokes (say at point $B$).

- Put differently, the opportunity cost of pizza increases as the number of pizzas increases.

- Why is that? Or, why is the production possibilities frontier bowed outward?

- It is because resources are not equally productive in all activities.

- When the economy uses most resources to produce lots of cokes and few pizzas (say point B), some resources best suited to produce pizzas are used to produce cokes, so some resources are likely not productive in producing cokes. From point $B$, increasing pizza by one unit will cause only a small reduction in the number of cokes.

- When the economy produces at point $E$, most resources are used to produce pizzas, to increase pizza by one unit will need to draw lots of resources from the coke industry (meaning giving up many cokes), the opportunity cost of pizza is high.

## 2. Allocative efficiency

- Production efficiency is achieved at each point of the production possibilities frontier.

- If combinations of goods on the production possibilities frontier provides the possibly greatest benefits, the economy then achieves the **allocative efficiency**.

- At which combination of outputs should the society produce?

- Or, at which point(s) on the production possibilities frontier achieve allocative efficiency?

  - It is determined by the **marginal cost** and **marginal benefit**.

### 2-1. Marginal cost

- The **marginal cost** of a good or service is the opportunity cost of producing **one more unit** of it.

- The opportunity cost is the number of cokes to give up in order to get Y million pizzas, while the marginal cost refers is the number of cokes to give up in order to get one additional pizza.

- Both must be referred to a particular point on the production possibilities frontier.

### 2-2. Marginal benefit

- Benefits from the combinations of goods are represented by **preferences** of consumers in the economy.

- Preferences are a description of a person's likes and dislikes.

- The **marginal benefit** of a good or service is the benefit received from consuming one more unit of it.

- For now, marginal benefit can be measured by the amount that a person is **willing to pay for an additional unit** of a good or service.

- Assumption on preference: economists assume that the benefit is increasing in the amount of goods consumed, for most goods.
  - It's called nonsatiation: the more we consume the higher the benefit.
  - There are exceptions, for example, potato, or probably any foods.
  - The Pizza and Coke example is unfortunate, let's assume that we they are never too much to digest.

- But, **marginal benefit** declines as the quantity of consumed goods goes higher. It is called the **principle of decreasing marginal benefit**.

- The more we have consumed a good, the smaller its marginal benefit to us, the less we are willing to pay for an additional unit of it.

### 2-3. Allocative efficiency: marginal benefit equals marginal cost

- When the economy cannot produce more of any one good without giving up some other good that the society values more highly, the economy has achieved **allocative efficiency**.

- When allocation is efficient, the economy is producing at the point on the PPF that the society prefers above all other points on the PPF.

- That occurs when the willingness to pay (marginal benefit) equals marginal cost.

## 3. Gains from trade

- Usually, people are good at producing certain goods but not good at producing other goods.

- If people only produce goods that they are good at, they gain by exchanging goods.

### 3-1. Absolute advantage and comparative advantage

- Formally, a person has an **absolute advantage** if that person is more productive than others.

- A person has a **comparative advantage** in an activity if that person can perform the activity at a lower opportunity cost than anyone else.

- Absolute advantage involves comparing productivities while comparative advantage involves comparing opportunity costs.

- Example: Joe and Liz who operate smoothie bars.

Joe | Liz

#### Liz
- In an hour, Liz can produce 30 smoothies or 30 salads.

- Liz's opportunity cost of producing 1 smoothie is 1 salad.

- Joe's opportunity cost of a salad is 1/5 smoothie. Liz's opportunity cost of a salad is 1 smoothie.

#### Liz
- Liz's opportunity cost of a smoothie is 1 salad. Joe's opportunity cost of a smoothie is 5 salads.

- Liz has a comparative advantage in producing smoothies, since her opportunity cost of a smoothie is less than Joe's.

## 4. Economic growth

- **Economic growth** refers to the expansion of production possibilities---an increase in the standard of living.

- Two key factors influence economic growth:
  - Technological change
  - Capital accumulation

- Technological change is the development of new goods and of better means of producing goods and services.

- Capital accumulation is the growth of capital resources, which includes human capital.

#### The Cost of Economic Growth

- To use resources in research and development and to produce new capital, we must decrease our production of consumption goods and services.

- Thus, economic growth is not free.

- The opportunity cost of economic growth is less current consumption.

## 5. Economic "Coordination"

- For the economy to reach efficient allocation, choices of individuals must be coordinated.

- To make coordination work, four social institutions have evolved:
  - Firms 
  - Markets
  - Property rights
  - Money

- A firm is an economic unit that hires factors of production and organizes those factors to produce.

- A market is an arrangement that enables buyers and sellers to get information and trade.

- Property rights are social arrangements that govern ownership, use, and disposal of resources, goods, or services.

- Money is a commodity or token that is legally acceptable as a means of payment.
