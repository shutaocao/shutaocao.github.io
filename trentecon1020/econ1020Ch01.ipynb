## 1. What is economics?

- **Economics** is the study of how individuals, businesses, governments, and the entire society make **choices**. These choices are influenced by resource **scarcity** and **incentives**.

- **Scarcity**: society has limited resources, and cannot produce all goods and services people _wish_ to have.

- An **incentive** is a reward or a penalty that induces an action.

### 1-1. Examples of economic choices

- Individuals choose
  - how much to consume and save, what to buy.
  - how many hours to work, and work for whom (self or others).
  - whether and what to invest their savings.
  
- Businesses make choices on
  - what to produce.
  - price setting.
  - investment.
  - hiring.
  - innovation.

- Governments decide
  - tax rates.
  - public spending.

### 1-2. Sub-fields of economics

- Two broad subfields: microeconomics and macroeconomics.

- **Microeconomics** is the study of choices that individuals and businesses make, the way those choices _interact_ in specific markets, and the influence of governments.

- **Macroeconomics** is the study of the performance of the national and global economies.

- The two subfields are distinct, because they address different questions.

- Microeconomics and macroeconomics are intertwined, increasingly closely.

- Other subfields fall into micro or macro or both: labor economics, industrial organization, game theory, international trade, ...


## 2.  Two big questions in economics

Two big questions summarize the scope of economics:

- How do choices end up determining _what_, _how_, and for _whom_ goods and services are produced?
  
- Do choices made in the pursuit of _self-interests_ also promote the _social interest_?

### 2-1. What, how, and for whom? The organization of production

- **What** to produce. In markets for goods and services, individuals are buyers, firms are sellers.
  - Nations do not all produce the same basket of goods and services.

- **How** to produce. Goods and services are produced by using **factors of production**
  - Land
  - Labour
  - Capital
  - (Entrepreneurship)

- In markets for the factors of production, households are sellers, firms are buyers.

- **Labour** input is the work time and work effort that people devote to producing goods and services.
  - Hours worked.
  - The quality of labour depends on **human capital**---the knowledge and skill that people obtain from education, on-the-job training, and work experience.

- Education attainment in Canada
<center> <img src="ParkinMacroIL/ParkinMacro_Ch01/image9.gif" align="middle" width=600 height=60 /></center>

- **Capital**: machinery and equipment; buildings and other structures (e.g., pipelines).

- **Entrepreneurship**: the human resource that organizes land, labour, and capital.
  - It is not a standard component of production factors.

- Other inputs of production are not called production factors. E.g., raw materials, energy, services.
  - Why? 1) They are themselves produced by production factors; 2) no value is added.

- Finally, **For Whom** refers to the final users of goods and services. Who receive goods and services depends on

  - Preference: people have preferences over goods and services.
  
  - Income: Land earns rent, labour earns wages, capital earns interest, entrepreneurship earns profit.

- Differences in returns to production factors are an recurring theme in economics: inequality.

  - Changes in inequality are often associated with technological progress.
  
  - Over the past three decades, there were important changes in inequality in Canada and globally.

### 2-2. Do choices for self-interest promote the social interest?

- Individuals make choices for their own best interests.

- Choices that are best for society as a whole are said to be in the social interest.

- Do choices for self-interest promote the social interest? Yes, in a market economy.

- In an economy with free markets, there are many buyers and sellers of numerous goods and services, all are primarily interested in their own well-being.

- The decentralized individual decision making and self-interests in a market economy maximize the well-being of society as a whole, in many cases.

#### Efficiency and equity

- Social interest has two dimensions: **efficiency** and **equity**.

- Efficiency versus equity is often a trade-off society faces.

- Resource use is **efficient** if it is impossible to make someone better off without making someone else worse off.

- Efficiency means that society gets the maximum benefits out of scarce resources.

- **Equity** is "justice according to natural law or right, specifically: freedom from bias or favoritism", according to Merriam-Webster.

- Simply put, equity is "fairness". It is to be more specific in order to have a meaning.
 - Is it fair for a low-skill worker to receive a **low** hourly wage rate, relative to a high-skill worker?
 - Is it fair that the top 1 percent population share 14 percent of total market income in Canada?
 - Equity = equality? Equality means benefits are uniformly distributed among society members.

## 3. The way of thinking in economics

### 3-1. Six key ideas/principles define the way of thinking in economics:

- A choice is a tradeoff.
- People make rational choices by comparing benefits and costs.
- Benefit is what you gain from something.
- Cost is what you must give up to get something.
- Most choices are "how-much" choices made at the margin.
- Choices respond to incentives.

### A choice is a tradeoff

- Every choice is a tradeoff because resources are scarce.

- Make an example of your own. Think of a choice you must make, what is the tradeoff?

### People are rational in making choices

- A rational choice is one that compares costs and benefits and achieves the greatest benefit over cost for the person making the choice, given the person's best knowledge and information.

- Only the wants of the person making a choice are relevant to determine its rationality.

- The idea of rational choice provides an answer to the first question: What goods and services will be produced and in what quantities? The answer is: Those that people rationally choose to buy.

### How do people choose rationally? 

- The answers turn on benefits and costs.

- Benefit: What people Gain.

- The benefit of something is the gain or pleasure that it brings and is determined by preferences

  - Preferences are what a person likes and dislikes and the intensity of those feelings.

### Cost: What people must give Up

- The **opportunity cost** of some thing is what you give up to get that thing.

- What is your opportunity cost of going to an AC/DC concert?

  - The things you can't afford to buy if you purchase the AC/DC ticket. 
  - The things you can't do with your time if you go to the concert.
  
- What is the opportunity cost of attending college for a talented athlete?

### Rational people make choices at the Margin

- For example, you want allocate some time between studying and messaging friends. 

- The choice is how many minutes to allocate to each activity.

- To make the choice, you compare the benefit of a little more study time with its cost: you make choice at the margin

- To that end, you evaluate the additional consequences of making incremental changes in the use of your time.

- The additional benefit from pursuing an incremental increase in an activity is its **marginal benefit**.

- The opportunity cost of pursuing an incremental increase in an activity is its **marginal cost**.

- If the marginal benefit from an incremental increase in an activity exceeds its marginal cost, your rational choice is to do more of that activity.

### Choices Respond to Incentives

- A change in marginal cost or a change in marginal benefit changes the incentives that we face and leads us to change our choice.

- A key idea of economics is that we can predict how choices will change by looking at changes in incentives.

- Incentives are also important for reconciling self-interest and the social interest.

### 3-2. Economist as social scientist

- Just as other scientists, economists explain the world by developing theory based on observations.

- Economists make assumptions to simply the complex world.

- More and new observations inspire improved and new theory.

- Economic models are frequently tested by comparing model predictions with the facts.

- **Positive analysis**: descriptive model of what the world `is'.

### 3-3. Economist as policy adviser

- Economic theory can help improve the world, as it is useful for policy analysis and policy recommendations.

- **Normative analysis**: prescriptive statements on what the world `ought to be'.

- Different models of how the world works can give conflicting policy recommendations.
  - Economists disagree, they may differ in scientific judgements.
  - Economists differ in values. Example: the definition or fairness can differ among economists.
